```python
with open('FanFics_Metadata_full.json', encoding='utf-8') as ff:
    fik = ff.read()
```

```python
fick_dicts = json.loads(fik)
```

```python
def prepare_one_row(fick):
    prepared_row = {}
    prepared_row['Title'] = fick['Title']
    prepared_row['Views'] = fick['Views']
    prepared_row['Readers'] = fick['Readers']
    prepared_row['Likes'] = fick['Likes']
    prepared_row['Collections'] = fick['Collections']
    prepared_row['Comments'] = fick['Comments']
    prepared_row['Genre'] = fick['Genre']
    prepared_row['Rating'] = fick['Rating']
    prepared_row['Size'] = fick['Size']['type']
    if fick['Dates'] is not None:
        prepared_row['Dates'] = fick['Dates']['published']
    for index, fandom in enumerate(fick['Fandom'][:3]):
        prepared_row[f'fandom_{index+1}'] = fandom['name']
    for index, pairing in enumerate(fick['Pairings'][:10]):
        for index2, person_name in enumerate(pairing['names']):
            prepared_row[f'pairing_{index+1}_{index2+1}'] = person_name
    for index, tag in enumerate(fick['Tags']):
        prepared_row[f'tag_{index+1}'] = tag
    for index, char in enumerate(fick['Characters'][:10]):
        prepared_row[f'char_{index+1}'] = char['name']
    if fick['Warnings'] is str:
        prepare_one_row['warning_1'] = fick['Warnings']
    elif fick['Warnings'] is not None:
        for index, warning in enumerate(fick['Warnings']):
            prepared_row[f'warning_{index+1}'] = warning
    return prepared_row
```

```python
prepare_one_row(fick_dicts[5])
```

```python
prepared_fick_dicts = [prepare_one_row(fick) for fick in fick_dicts]
```

```python
import pandas as pd

hp = pd.DataFrame(prepared_fick_dicts)
```
